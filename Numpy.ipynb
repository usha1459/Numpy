```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\\Codegnan\\Data Science\\Class\\Numpy\\Indian_Cricket_Team_Dataset.csv")
```

```python
df1 = df.copy()
```

```python
df1.info()
```

## Query 1: Convert 'Age' column to a NumPy array.

```python
# Convert 'Age' column to a NumPy array
age_array = df1['Age'].to_numpy()

print(age_array)  # Output: [25 30 35]
```

## Query 2: Find the average age of players.

```python
df1.describe()
```

```python
df1['Age'].mean()
```

## Query 3: Find the oldest and youngest player age.

```python
df1['Age'].max()
```

```python
df1['Age'].min()
```

## Query 4: Count the total number of players above 30 years old

```python
df1[df1['Age']>30]['Age'].count()
```

## Query 5: Find the median batting average.

```python
df1['Batting Average'].median()
```

## Query 6: Find the standard deviation of strike rates.

```python
df1['Batting Average'].std()
```

## Query 7: Find the 90th percentile of batting average.

```python
df1['Batting Average'].quantile(0.90)
```

## Query 8: Find the number of players with strike rate above 150.

```python
df1[df1['Strike Rate']>150]['Strike Rate'].count()
```

## Query 9: Reshape the first 20 batting averages into a 4x5 matrix.

```python
avg_20 = df1['Batting Average'].head(20).to_numpy()
avg_20_matrix = avg_20.reshape(4,5)
display(avg_20_matrix)
```

## Query 10: Compute row-wise sum of reshaped matrix.

```python
row_sum = avg_20_matrix.sum(axis = 1)
print(row_sum)
```

## Query 11: Compute row-wise sum of reshaped matrix.

```python
column_sum = avg_20_matrix.sum(axis = 0)
print(column_sum)
```

## Query 12: Transpose the reshaped matrix.

## Query 13: Find the variance of batting average

```python
var = df1['Batting Average'].var()
print(var)
```

## Query 14: Stack age and matches played horizontally

```python
a = df1['Age']
b = df1['Matches Played']
stacked = np.stack((a,b),axis = 1)
print(stacked)
```

## Query 15: Split the batting average array into 5 equal parts.

```python
a = df1['Batting Average'].to_numpy()
split = np.array_split(a,5)
print(split)
```
